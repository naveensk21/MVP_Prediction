```python
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
years = list(range(1995, 2024))
```

```python
import requests 
import time

for year in years:
    url = f"https://www.basketball-reference.com/awards/awards_{year}.html"
    data = requests.get(url)
    
    # w+ to overwrite existing files
    with open("mvp_data/{}.html".format(year), "w+") as fp:
        fp.write(data.text)
    print("extracted {}".format(year))
    time.sleep(10)
    
```

```python
from bs4 import BeautifulSoup
import pandas as pd

# read single file
with open("mvp_data/1995.html") as fp:
    page = fp.read()

# intialize bs4
soup = BeautifulSoup(page, "html.parser")

# remove the over header row
soup.find('tr', class_='over_header').decompose()

# find mvp table
mvp_table = soup.find(id='mvp')

# read the file in pandas
mvp_1995 = pd.read_html(str(mvp_table))[0]

mvp_1995
```

```python
# for all extracted data
dfs = []
for year in years:
    with open("mvp_data/{}.html".format(year)) as fp:
        page = fp.read()
    
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_='over_header').decompose()
    mvp_table = soup.find(id='mvp')
    
    each_mvp = pd.read_html(str(mvp_table))[0]
    
    each_mvp['Year'] = year
    
    dfs.append(each_mvp)  
```

```python
# combine the df
combined_mvps = pd.concat(dfs)
```

```python
from selenium import webdriver

years = list(range(1996, 2025))

driver = webdriver.Chrome()

for year in years:
    players_stats_url = f"https://www.basketball-reference.com/leagues/NBA_{year}_per_game.html"
    driver.get(players_stats_url)
    
    # to scroll down through the page to load all the tables to extract
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    extracted_player_stats = driver.page_source
    
    with open('all_players_data/{}.html'.format(year), 'w+') as fp:
        fp.write(extracted_player_stats)
    
    print(f"extracted player statistics of year-{year}...")
    time.sleep(5)

driver.close()
```

```python
years = list(range(1995, 2025))

dfs = []

for year in years:
    with open('all_players_data/{}.html'.format(year)) as fp:
        all_players_page = fp.read()

    soup = BeautifulSoup(all_players_page, 'html.parser')

    soup.find('tr', class_='thead').decompose

    all_players_table = soup.find(id='per_game_stats')

    all_player_stats = pd.read_html(str(all_players_table))[0]

    all_player_stats['Year'] = year

    dfs.append(all_player_stats)
```

```python
combined_all_player_stats = pd.concat(dfs)
```

```python
team_stats_url = f'https://www.basketball-reference.com/leagues/NBA_1995_standings.html'

data = requests.get(team_stats_url)

with open('team_data/1995.html', 'w+') as fp:
    fp.write(data.text)

print(f'Extracted team data for year 1995...')
```

```python
# extract the wester and easter conference standing into a dataframe
years = list(range(1995, 2025))
dfs = []
for year in years:
    with open('team_data/{}.html'.format(year)) as fp:
        page = fp.read()

    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_='thead').decompose
    east_team_table = soup.find(id='divs_standings_E')
    team = pd.read_html(str(east_team_table))[0]
    team['Year'] = year
    team['Team'] = team['Eastern Conference']
    del team['Eastern Conference']
    dfs.append(team)


    soup = BeautifulSoup(page, 'html.parser')
    soup.find('tr', class_='thead').decompose
    west_team_table = soup.find(id='divs_standings_W')
    team = pd.read_html(str(west_team_table))[0]
    team['Year'] = year
    team['Team'] = team['Western Conference']
    del team['Western Conference']
    dfs.append(team)
```

```python
combined_teams = pd.concat(dfs)
```

```python
combined_teams.to_csv('combined_teams.csv')
```
