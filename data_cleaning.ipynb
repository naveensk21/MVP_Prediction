```python
# read the mvp_data
import pandas as pd 

mvp_data = pd.read_csv('combined_mvp.csv')
```

```python
# select the necessary columns/ duplicate coloumns in all_players dataset 
mvp_data = mvp_data[['Player', 'Year', 'Pts Won', 'Pts Max', 'Share']]
mvp_data
```

```python
all_players = pd.read_csv('combined_all_players.csv')
```

```python
all_players
```

```python
del all_players['Unnamed: 0']
```

```python
all_players.head(50)
```

```python
# remove the asterix in the players names
all_players['Player'] = all_players['Player'].str.replace('*', '')
```

```python
all_players['Player'].head(51)
```

```python
# remove row with player and no stats
# mask = all_players['Player'] == 'Player'

# all_players = all_players[~mask]

all_players = all_players.drop(all_players[all_players['Player'] == 'Player'].index)

all_players.head(50)
```

```python
# reset the index
all_players.reset_index(inplace=True, drop=True)
all_players.head(50)
```

```python
# some players have several stats as they player for different teams 
all_players.groupby(['Player', 'Year']).get_group(('Willie Anderson',1996))
```

```python
def single_player(df):
    if df.shape[0] == 1:
        return df
    # return the row with players total stats TOT
    else:
        row = df[df['Tm'] == 'TOT']
        # replace the TOT with the last team the player played 
        row['Tm'] = df.iloc[-1,:]['Tm']
        return row

all_players = all_players.groupby(['Player', 'Year']).apply(single_player)       
```

```python
all_players.head(25)
```

```python
# drop two index 
all_players.index = all_players.index.droplevel()
```

```python
all_players.index = all_players.index.droplevel()
```

```python
# merge the mvp and all_players dataset
# merge on the player name 
# outer merge as the mvp player does not have all the players name 
combined_data = all_players.merge(mvp_data, how='outer', on=['Player', 'Year'])
```

```python
# replace NaN with zeros
combined_data[['Pts Won', 'Pts Max', 'Share']] = combined_data[['Pts Won', 'Pts Max', 'Share']].fillna(0)
```
